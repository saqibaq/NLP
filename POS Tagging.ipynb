```
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```
# News article
txt = 'A top scientist from the Department of Homeland Security said Thursday that preliminary research shows that the coronavirus dies out more quickly in hot, humid weather than in colder, drier conditions. “Increasing the temperature and humidity … is generally less favorable to the virus,” said William Bryan, head of Science and Technology at DHS, said during the daily Coronavirus Task Force daily briefing at the White House. (Watch Mr. Bryan’s remarks under “Resources” below.) Bryan said the preliminary study showed that the half life of the virus was lower at higher temperatures and humidity than in colder weather.'
```

```
punc = [',', '.', '“', '…', '”', '(', '’', ')']
words = []  # List of words after removing punctuations
word_tag = []  # List of tuples containing word and its tag
tokenized = sent_tokenize(txt) 
for i in tokenized: 
      
    wordsList = nltk.word_tokenize(i) 
    word_preprocess = [word for word in wordsList if word not in punc]
    words = words + word_preprocess

    tagged = nltk.pos_tag(word_preprocess)
    word_tag = word_tag + tagged
    
```

```
all_tags = set() # All unique tags identified in the article
for i,j in word_tag:
  all_tags.add(j)
```

```
tag_count = dict() # Dictionary whose key is tag and value is number of times it appeared
for i in all_tags:
  for a,b in  word_tag:
    if b==i:
      tag_count[i] = tag_count.get(i,0) + 1 
```

```
from collections import defaultdict
tag_with_all_words = defaultdict(int)  #Dictionary whose key is tag and value is a list of words with this tag
for i in all_tags:
  for a,b in  word_tag:
    if b==i:
      if tag_with_all_words[i] == 0:
        tag_with_all_words[i] = [a]
      else:
        tag_with_all_words[i] = tag_with_all_words[i] + [a]
```
